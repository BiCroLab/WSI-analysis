```python
import plotly.graph_objects as go
import networkx as nx
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import sparse
sns.set()
import glob, os
import matplotlib.pyplot as plt
```

```python
pattern="*.txt.cc"

os.chdir("/media/garner1/hdd2/DAPI/old")
fig = plt.figure(figsize=(20,12))
for file in glob.glob(pattern):
    print(file)
    cc = np.loadtxt(file)
    sns.distplot(cc,hist=False,label=file)
labels1=[file for file in glob.glob(pattern)]

os.chdir("/media/garner1/hdd2/ND2/segmentation")
for file in glob.glob(pattern):
    print(file)
    cc = np.loadtxt(file)
    sns.distplot(cc,hist=False,label=file)
labels2=[file for file in glob.glob(pattern)]
```
